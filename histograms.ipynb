```python
t = [1,2,3,1,2,3,1,1,2,6,7,9,2,6,2,6,7,2,3,2,6,2,2]
```

```python
hist = {}
for x in t:
    hist[x] = hist.get(x,0) +1
print(hist)
print('value to frequency')
```

```python
n = float(len(t))
print(n)
pmf = {}
for x, freq in hist.items():
    pmf[x] = freq/n
print('pmf')
print(pmf)
```

```python
import Pmf
ist = Pmf.MakeHistFromList(t)
ist.Freq(2)
ist.Values()
```

```python
import matplotlib.pyplot as pyplot
pyplot.pie(t)
pyplot.show()
```
